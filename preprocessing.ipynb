```python
images_path = 'data/train/train_images'
masks_path = 'data/train/train_masks'
random_image_and_mask(images_path, masks_path)
```

```python
# Function that will split the images and masks into training and testing sets according to the percentage given and then move them into their respective folders
def train_test_split(images_path, masks_path, percentage):
    images_dir = os.listdir(images_path)
    masks_dir = os.listdir(masks_path)
    # Make sure that there are the same amount of images and masks
    assert len(images_dir) == len(masks_dir)
    # Get the number of images
    n_images = len(images_dir)
    # Calculate the number of images for the testing set
    n_test_images = int(percentage * n_images)
    # Randomly choose the images that will be used for the testing set
    test_images = np.random.choice(images_dir, n_test_images, replace=False)
    # Move the images and masks to their respective folders
    for image in test_images:
        os.replace(os.path.join(images_path, image), os.path.join('test_images', image))
        os.replace(os.path.join(masks_path, image.replace('jpg', 'png')), os.path.join('test_masks', image.replace('jpg', 'png')))
```

```python
train_test_split(images_path, masks_path, 0.2)
```
